```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student-mat.csv")
```

```python
df.head()
```

```python
print("Average Final Grade:",df['G3'].mean())
```

```python
plt.hist(df['G3'], bins = 10, color='skyblue', edgecolor='black')
plt.title("Distribution of final grades")
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['studytime'], df['G3'])
plt.xlabel('Study Time')
plt.ylabel('G3')
plt.title('Study Time vs Performance')
plt.show()
```

```python
df.groupby('sex')['G3'].mean()
```

```python
import matplotlib.pyplot as plt

df.groupby('sex')['G3'].mean().plot(kind='bar')
plt.ylabel('Average G3 Score')
plt.title('Average Final Grade (G3) by Gender')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['G3'], bins=10, edgecolor='black')
plt.xlabel('G3 Grade')
plt.ylabel('Number of Students')
plt.title('Distribution of Final Grades (G3)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=df, x='G3', hue='sex', bins=10, kde=True)
plt.title('G3 Distribution by Gender')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['studytime'], df['G3'], alpha=0.6)
plt.xlabel('Study time')
plt.ylabel('Final Grade (G3)')
plt.title('Study Time vs Final Grade')
plt.show()
```

```python
avg_scores = df.groupby('sex')['G3'].mean()

avg_scores.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Average G3 Score')
plt.title('Average Final Grade by Gender')
plt.show()
```
